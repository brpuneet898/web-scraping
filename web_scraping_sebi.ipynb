# Extracting the href links from the HTML Page

```python
import chromedriver_autoinstaller
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

chromedriver_autoinstaller.install()

chrome_binary = '/usr/bin/google-chrome-stable'

chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.binary_location = chrome_binary

service = Service(chromedriver_autoinstaller.install())
driver = webdriver.Chrome(service=service, options=chrome_options)

url = "https://www.sebi.gov.in/section-search.html?searchval=insider%20trading&searchcontext=Orders&searchfromdate=01-01-2025&searchtodate=30-01-2025"
driver.get(url)

driver.implicitly_wait(5)

links = driver.find_elements(By.TAG_NAME, 'a')
link_urls = [link.get_attribute('href') for link in links if link.get_attribute('href')]

print(f"Extracted {len(link_urls)} links:")
for link in link_urls:
    print(link)

driver.quit()
```

# Extract the relevant links from all the links.

```python
import re

chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.binary_location = chrome_binary

service = Service(chromedriver_autoinstaller.install())
driver = webdriver.Chrome(service=service, options=chrome_options)

url = "https://www.sebi.gov.in/section-search.html?searchval=insider%20trading&searchcontext=Orders&searchfromdate=01-01-2025&searchtodate=30-01-2025"
driver.get(url)

driver.implicitly_wait(5)

links = driver.find_elements(By.TAG_NAME, 'a')
link_urls = [link.get_attribute('href') for link in links if link.get_attribute('href')]

pattern = r"https://www\.sebi\.gov\.in/enforcement/orders/.*"

filtered_links = [link for link in link_urls if re.match(pattern, link)]

unique_filtered_links = list(set(filtered_links))

print(f"Extracted {len(unique_filtered_links)} unique matching links:")
for link in unique_filtered_links:
    print(link)

driver.quit()
```

# Find the iframe from each links, in which the PDF file will be present.

```python
import time
import chromedriver_autoinstaller
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

chromedriver_autoinstaller.install()

chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(options=chrome_options)

iframe_src_list = []

def check_iframe_presence(url):
    try:
        driver.get(url)
        time.sleep(3)
        iframes = driver.find_elements(By.TAG_NAME, 'iframe')
        if len(iframes) > 0:
            print(f"Iframe found for {url}. Total iframes: {len(iframes)}")
            for iframe in iframes:
                iframe_src = iframe.get_attribute('src')
                iframe_src_list.append(iframe_src)
        else:
            print(f"No iframe found for {url}")
    except Exception as e:
        print(f"Error checking iframe for {url}: {e}")
for link in unique_filtered_links:
    check_iframe_presence(link)

unique_iframe_sources = list(set(iframe_src_list))

print("Unique Iframe Sources:")
for iframe in unique_iframe_sources:
    print(iframe)

driver.quit()
```

```python
print(unique_iframe_sources)
```

# Extract the links from iframe

```python
filtered_iframe_sources = [src for src in unique_iframe_sources if 'https://www.sebi.gov.in/web/?file=https://www.sebi.gov.in/sebi_data/attachdocs/mar-2025' in src]

print("Filtered Iframe Sources (PDF links):")
for iframe in filtered_iframe_sources:
    print(iframe)
```

# Once the PDF links are ready, download the PDFs -> save it in a zip file.

```python
import requests
import zipfile
import os
from google.colab import files

def download_pdf(pdf_url, save_folder):
    try:
        actual_pdf_url = pdf_url.split("file=")[1]
        response = requests.get(actual_pdf_url)
        if response.status_code == 200:
            pdf_filename = actual_pdf_url.split('/')[-1]
            pdf_path = os.path.join(save_folder, pdf_filename)
            with open(pdf_path, 'wb') as pdf_file:
                pdf_file.write(response.content)

            print(f"Successfully downloaded: {pdf_filename}")
            return pdf_path
        else:
            print(f"Failed to download PDF: {pdf_url} (Status code: {response.status_code})")
            return None
    except Exception as e:
        print(f"Error downloading PDF from {pdf_url}: {e}")
        return None

save_folder = '/content/downloaded_pdfs'
os.makedirs(save_folder, exist_ok=True)

pdf_files = []

for pdf_link in filtered_iframe_sources:
    pdf_path = download_pdf(pdf_link, save_folder)
    if pdf_path:
        pdf_files.append(pdf_path)

zip_filename = '/content/pdfs.zip'
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for pdf_file in pdf_files:
        zipf.write(pdf_file, os.path.basename(pdf_file))

files.download(zip_filename)

print(f"All PDFs have been zipped and are ready for download: {zip_filename}")
```
